```python
import pandas as pd
import numpy as np

# Load the data
csvData2 = pd.read_csv("updated_2019-2024_droppedRows_and_columns.csv")
df2 = pd.DataFrame(csvData2)

# Create a dictionary to store the profCount, assoCount, and docCount values for each idOSYM
count_dict = {}


count_dict2 = {}



def find_recent_counts(idOSYM, academicYear):
    if idOSYM in count_dict:
        for year in sorted(count_dict[idOSYM].keys(), reverse=True):
            if year < academicYear:
                counts = count_dict[idOSYM][year]
                if counts != (0, 0, 0) and not pd.isna(counts[0]) and not pd.isna(counts[1]) and not pd.isna(counts[2]):
                    return counts
    return None

# Function to find the closest non-zero and non-empty values for a given idOSYM and academicYear
def find_sameDep_counts(idOSYM, academicYear, universityName, departmentName,scholarshipRate):
    rates = ["100", "75", "50", "25","0",""]
    for rate in rates:
        if scholarshipRate != rate and (universityName,departmentName,academicYear,rate) in count_dict2:
            if count_dict2[(universityName,departmentName,academicYear,rate)] != (0, 0, 0) and not pd.isna(count_dict2[(universityName,departmentName,academicYear,rate)][0]) and not pd.isna(count_dict2[(universityName,departmentName,academicYear,rate)][1]) and not pd.isna(count_dict2[(universityName,departmentName,academicYear,rate)][2]):
                return count_dict2[(universityName,departmentName,academicYear,rate)]

def find_closest_counts(idOSYM, academicYear):
    if idOSYM in count_dict:
        years = sorted(count_dict[idOSYM].keys())
        closest_year = None
        for year in years:
            if year != academicYear:
                counts = count_dict[idOSYM][year]
                if counts != (0, 0, 0) and not pd.isna(counts[0]) and not pd.isna(counts[1]) and not pd.isna(counts[2]):
                    if closest_year is None or abs(year - academicYear) < abs(closest_year - academicYear):
                        closest_year = year
        if closest_year is not None:
            return count_dict[idOSYM][closest_year]
    return None
    
    # If no closest year found in count_dict, check count_dict2
    
    return None

for i in range(0, 4):
    
    count_dict = {}

    # Iterate through the DataFrame to populate the dictionary with non-zero and non-empty values
    for index, row in df2.iterrows():
        key = row["idOSYM"]
        if key not in count_dict:
            count_dict[key] = {}
        if row["academicYear"] not in count_dict[key]:
            count_dict[key][row["academicYear"]] = (row["profCount"], row["assoCount"], row["docCount"])

    count_dict2 = {}

    for index, row in df2.iterrows():
        key = (row["universityName"],row["departmentName"],row["academicYear"],row["scholarshipRate"])
        if key not in count_dict2:
            count_dict2[key] = (row["profCount"], row["assoCount"], row["docCount"])
    
    # Update the profCount, assoCount, and docCount values in df2
    for index, row in df2.iterrows():
        if (pd.isna(row["profCount"]) and pd.isna(row["assoCount"]) and pd.isna(row["docCount"])) or (row["profCount"] == "" and row["assoCount"] == "" and row["docCount"] == "") or (row["profCount"] == 0 and row["assoCount"] == 0 and row["docCount"] == 0):
            closest_counts = find_sameDep_counts(row["idOSYM"], row["academicYear"], row["universityName"], row["departmentName"],row["scholarshipRate"])
            if closest_counts:
                df2.at[index, "profCount"] = closest_counts[0]
                df2.at[index, "assoCount"] = closest_counts[1]
                df2.at[index, "docCount"] = closest_counts[2]
    
    for index, row in df2.iterrows():
        if row["profCount"] == 0 and row["assoCount"] == 0 and row["docCount"] == 0:
            recent_counts = find_recent_counts(row["idOSYM"], row["academicYear"])
            if recent_counts:
                df2.at[index, "profCount"] = recent_counts[0]
                df2.at[index, "assoCount"] = recent_counts[1]
                df2.at[index, "docCount"] = recent_counts[2]
    
    for index, row in df2.iterrows():
        if (row["profCount"] == 0 and row["assoCount"] == 0 and row["docCount"] == 0)  or (pd.isna(row["profCount"]) and pd.isna(row["assoCount"]) and pd.isna(row["docCount"])) or (row["profCount"] == "" and row["assoCount"] == "" and row["docCount"] == ""):
            closest_counts = find_closest_counts(row["idOSYM"], row["academicYear"])
            if closest_counts:
                df2.at[index, "profCount"] = closest_counts[0]
                df2.at[index, "assoCount"] = closest_counts[1]
                df2.at[index, "docCount"] = closest_counts[2]
        
    
    

# df2["profCount"] = df2["profCount"].fillna(0).astype(int)
# df2["assoCount"] = df2["assoCount"].fillna(0).astype(int)
# df2["docCount"] = df2["docCount"].fillna(0).astype(int)
# Save the updated DataFrame to a new CSV file
df2.to_csv("updated_2019-2024_with_filled_counts3.csv", index=False, encoding="utf-8")
```

```python
csvData17 = pd.read_csv("updated_2019-2024_near_near_final.csv")
df17 = pd.DataFrame(csvData17)


def update_rankings(row):
    if row["occupiedSlots"] != 0:
        if pd.isna(row["baseScore"]) or row["baseScore"] == "":
            row["baseRanking"] = "0"
        if pd.isna(row["topScore"]) or row["topScore"] == "":
            row["topRanking"] = "0"
    return row
df17 = df17.apply(update_rankings, axis=1)

meanBaseScore = {}
meanTopScore = {}

idOSYM = 0
count = 0
for index, row in df17.iterrows():
    if idOSYM != row["idOSYM"]:
        if idOSYM != 0:
            meanBaseScore[idOSYM] = float(meanBaseScore[idOSYM]) / count
        idOSYM = row["idOSYM"]
        key = row["idOSYM"]
        if key not in meanBaseScore:
            meanBaseScore[key] = float(row["baseScore"])
        count = 1
    else:
        count += 1
        key = row["idOSYM"]
        if key in meanBaseScore:
            meanBaseScore[key] += float(row["baseScore"])
meanBaseScore[idOSYM] = float(meanBaseScore[idOSYM]) / count

idOSYM = 0
count = 0
for index, row in df17.iterrows():
    if idOSYM != row["idOSYM"]:
        if idOSYM != 0:
            meanTopScore[idOSYM] = float(meanTopScore[idOSYM]) / count
        idOSYM = row["idOSYM"]
        key = row["idOSYM"]
        if key not in meanTopScore:
            meanTopScore[key] = float(row["topScore"])
        count = 1
    else:
        count += 1
        key = row["idOSYM"]
        if key in meanTopScore:
            meanTopScore[key] += float(row["topScore"])
meanTopScore[idOSYM] = float(meanTopScore[idOSYM]) / count


for index,row in df17.iterrows():
    if row["baseScore"] == "0" or row["baseScore"] == 0:
        closest_counts = meanBaseScore[row["idOSYM"]]
        if closest_counts:
            df17.at[index, "baseScore"] = closest_counts
    if row["topScore"] == "0" or row["topScore"] == 0:
        closest_counts = meanTopScore[row["idOSYM"]]
        if closest_counts:
            df17.at[index, "topScore"] = closest_counts


df17.to_csv("updated_2019-2024_JSUTBEFOREFINAL.csv", index=False, encoding="utf-8")

```
