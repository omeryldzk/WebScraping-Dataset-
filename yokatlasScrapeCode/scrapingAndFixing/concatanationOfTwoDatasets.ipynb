```python
import pandas as pd

# Load the data and specify the 'scholarshipRate' and 'idOSYM' columns as strings
csvData4 = pd.read_csv("test.csv", encoding="utf-8", dtype={"scholarshipRate": str, "idOSYM": str})
dataFrame2 = pd.DataFrame(csvData4)

# Create a dictionary to store the idOSYM values for non-empty rows
idOSYM_dict = {}



# Populate the dictionary with non-empty idOSYM values
for index, row in dataFrame2.iterrows():
    key = ( row['universityName'], row['faculty'], row['departmentName'],row["language"], row['scholarshipRate'])
    if pd.notna(row['idOSYM']) and row['idOSYM'].strip() != '':
        idOSYM_dict[key] = row['idOSYM']

# Update the idOSYM for rows where it is empty
for index, row in dataFrame2.iterrows():
    key = ( row['universityName'], row['faculty'], row['departmentName'],row["language"], row['scholarshipRate'])
    if pd.isna(row['idOSYM']) or row['idOSYM'].strip() == '':
        if key in idOSYM_dict:
            dataFrame2.at[index, 'idOSYM'] = idOSYM_dict[key]

idOSYM_dict2 = {}

# Populate the dictionary with non-empty idOSYM values
for index, row in dataFrame2.iterrows():
    key = ( row['universityName'], row['departmentName'],row["language"], row['scholarshipRate'])
    if pd.notna(row['idOSYM']) and row['idOSYM'].strip() != '':
        idOSYM_dict2[key] = row['idOSYM']

# Update the idOSYM for rows where it is empty
for index, row in dataFrame2.iterrows():
    key = ( row['universityName'], row['departmentName'],row["language"], row['scholarshipRate'])
    if pd.isna(row['idOSYM']) or row['idOSYM'].strip() == '':
        if key in idOSYM_dict2:
            dataFrame2.at[index, 'idOSYM'] = idOSYM_dict2[key]

# Save the updated DataFrame to a new CSV file
dataFrame2.to_csv("updated_2019-2024(withIDs).csv", index=False, encoding="utf-8")
```
